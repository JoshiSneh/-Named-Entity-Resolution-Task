```python
import re

def pre_processing(text):
    '''
    
    This function normalize the input text by:
    1) Lower casing the text.
    2) Removing the special characters
    3) Removing the extra white spaces
    
    INPUT: string: text
    OUTPUT: string: normalize text
    
    '''
    text = str(text).lower()
    text = re.sub(r'[^a-z0-9]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
'''
Applying the pre_processing function to the below column of the Amazon dataset.
'''
df_amazon['product_name'] = df_amazon['product_name'].apply(pre_processing)
df_amazon['brand'] = df_amazon['brand'].apply(pre_processing)
df_amazon['description'] = df_amazon['description'].apply(pre_processing)
```

```python
'''
Applying the pre_processing function to the below column of the Flipkart dataset.
'''
df_flipkart['product_name'] = df_flipkart['product_name'].apply(pre_processing)
df_flipkart['brand'] = df_flipkart['brand'].apply(pre_processing)
df_flipkart['description'] = df_flipkart['description'].apply(pre_processing)
```

```python
'''
Changing the data type of retail_price to int
'''
df_amazon['retail_price'] = pd.to_numeric(df_amazon['retail_price'], errors='coerce').astype(int)
df_flipkart['retail_price'] = pd.to_numeric(df_flipkart['retail_price'], errors='coerce').astype(int)
```

```python
import re

def normalize_category(category_tree):
    '''
    
    This function normalize the category column by normalizing the values of the category and extracting the main category from the category tree
    
    INPUT: string: category_tree
    OUTPUT: string: normalize_main_category
    
    '''
    if pd.isna(category_tree):
        return ''

    category_tree = re.sub(r'[\[\]\']', '', category_tree)
    categories = category_tree.split('>>')
    categories = [category.strip().lower() for category in categories]

    normalized_category =  ' >> '.join(categories)

    if not normalized_category:
        return ''
    
    main_category = normalized_category.split(' >> ')[0].split('"')[1]

    return main_category
```

```python
'''
Applying the normalize_category function to product_category_tree column
'''
df_amazon["product_category_tree"] = df_amazon["product_category_tree"].apply(normalize_category)
df_flipkart["product_category_tree"] = df_flipkart["product_category_tree"].apply(normalize_category)
```

```python
'''
Importing fuzzywuzzy library which is used for the fuzzy matching and for the string matching.
'''

from fuzzywuzzy import process
```

```python
def matching_products(amazon, flipkart):
    
    '''
    
    This function is the main algorithm for the product/entity matching
    
    INPUT: datafram: amazon, flipkart
    OUTPUT: list: matched_products
    
    '''
    matched_products = []
    
    for index, row in amazon.iterrows():

        amazon_product_description = row["product_description"]
        amazon_brand_name = row['brand']
        amazon_retail_price = row['retail_price']
        amazon_category = row['product_category_tree']
        
        filtered_flipkart = flipkart[(amazon_brand_name == flipkart['brand']) & (flipkart['retail_price'].between(amazon_retail_price * 0.9, amazon_retail_price * 1.1)) & (amazon_category == flipkart["product_category_tree"])]

        # print(filtered_flipkart.shape)
        if filtered_flipkart.shape[0] > 0:

            flipkart_match_data = process.extractOne(amazon_product_description, filtered_flipkart['product_description'])

            # print(flipkart_match_data[2])
            if flipkart_match_data and flipkart_match_data[1] == 100: 
                flipkart_match = flipkart.iloc[flipkart_match_data[2]]
                matched_products.append((row['uniq_id'], row['product_name'],row["retail_price"],flipkart_match['uniq_id'],flipkart_match['product_name'],flipkart_match['retail_price']))

    return matched_products
```

```python
matched_products = matching_products(df_amazon,df_flipkart)
# matched_products
```
